```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
dt = pd.read_csv('iris.csv')
```

```python
dt.shape
```

```python
dt.head()
```

```python
dt.plot(x='sepal_length', y='sepal_width')
```

```python
dt.plot(x='sepal_length', y='sepal_width', kind='scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
dt.plot(y='sepal_length', kind='box')
plt.ylabel('sepal width (cm)')
```

```python
dt.plot(y='sepal_length', kind='hist')
plt.xlabel('sepal length (cm)')
```

```python
dt.plot(y='sepal_length', kind='hist', bins=30, range=(4,8), normed=True)
plt.xlabel('sepal length (cm)')
plt.show()
```

```python
dt.plot(y='sepal_length', kind='hist', bins=30, range=(4,8), cumulative=True, normed=True)
plt.xlabel('sepal length (cm)')
plt.title('Cumulative distribution function (CDF)')
```
